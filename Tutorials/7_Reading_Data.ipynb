```python
# The BTC market price file does not have a header. 
# Pandas assign the first row as the header. 
# How to suppress this behavior? 

df = pd.read_csv(btc_file_url, header= None)
df.head()
```

```python
# Identify the null values for empty strings '', ? and -

df = pd.read_csv(btc_file_url, 
                 header= None, 
                 na_values=['', '?', '-'])
df.head()
```

```python
# Add column names

df = pd.read_csv(btc_file_url, 
                 header= None, 
                 na_values=['', '?', '-'],
                 names=['Timestamp', 'Price'])
df.head()
```

```python
# Force pandas to use certain dtype

df = pd.read_csv(btc_file_url, 
                 header= None, 
                 na_values=['', '?', '-'],
                 names=['Timestamp', 'Price'],
                 dtype={'Price': 'float'})
df.head()
```

```python
df.dtypes
```

```python
# Timestamp is a string
# Parse the Timestamp column to Datetime objects using to_datetime method

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.head()
```

```python
df.dtypes
```

```python
# Or using parse_dates parameter of Pandas read_csv

df = pd.read_csv(btc_file_url,
                 header=None,
                 na_values=['', '?', '-'],
                 names=['Timestamp', 'Price'],
                 dtype={'Price': 'float'},
                 parse_dates=[0])

df.head()
```

```python
df.dtypes
```

```python
# Choose Timstamp as the index

df = pd.read_csv(btc_file_url,
                 header=None,
                 na_values=['', '?', '-'],
                 names=['Timestamp', 'Price'],
                 dtype={'Price': 'float'},
                 parse_dates=[0],
                 index_col=[0] # Timestamp has col index 0
                 )

df.head()
```

```python
exam_df[['math_score', 'french_score']].dtypes
```

```python
# Specify which columns to read

exam_df = pd.read_csv(exam_file,
                      sep='>',
                      decimal=',',
                      usecols=['first_name', 'last_name', 'age'] # Specify col name
                      )
exam_df.head()
```

```python
# ExcelFile will let us work with sheets easily, and will be faster than the previous read_excel method.
excel_file = pd.ExcelFile('products.xlsx')
excel_file.sheet_names
```

## Save to Excel File

```python
products.to_excel('out.xlsx')
pd.read_excel('out.xlsx').head()
```

```python
# Specify the sheet name
# Further calls of to_excel with different sheet names will only overwrite the first sheet instead of adding additional sheets
products.to_excel('out.xlsx',
                  sheet_name='Products')
```

```python
# Reemoving the index also removes the data associated with it
products.to_excel('out.xlsx',
                  index=None)
pd.read_excel('out.xlsx').head()
```

```python
# Position data with startrow and startcol
products.to_excel('out.xlsx',
                  sheet_name='Products',
                  startrow=1,
                  startcol=2)
```

```python
# To save multiple sheets we need to create a writer

writer = pd.ExcelWriter('out.xlsx')
writer
```

```python
with writer:
    products.to_excel(writer, sheet_name='Products')

pd.read_excel('out.xlsx', sheet_name='Products').head()
```

```python
# Add Merchants sheet to the same file using writer
with writer:
    merchants.to_excel(writer, sheet_name='Merchants')

pd.read_excel('out.xlsx', sheet_name='Products').head()
```

```python
pd.read_excel('out.xlsx', sheet_name='Merchants').head()
```

```python
# Save multiple sheets at the same time
with pd.ExcelWriter('out.xlsx') as writer:
    products.to_excel(writer, sheet_name='Products')
    merchants.to_excel(writer, sheet_name='Merchants')
```

```python
import pandas as pd
import sqlite3
```

```python
# Or use Pandas to read the data from the Database

# we still need to connect to the database
conn = sqlite3.connect('/content/chinook.db')

# Then we can read straight to a dataframe using pandas
df = pd.read_sql('SELECT * FROM employees;', conn)
df.head()
```

### Use requests to get HTML code from an URL

```python
import requests

html_url = "https://en.wikipedia.org/wiki/The_Simpsons"
r = requests.get(html_url)

# The given URL we can see multiple tables about The Simpsons TV show.
wiki_tables = pd.read_html(r.text, header=0)
len(wiki_tables) # there are many tables, but only one is visible
```

```python
simpsons = wiki_tables[1]
simpsons.head()
```

```python
# Quick clean on the table: remove extra header rows and set Season as index.
simpsons.drop([0, 1], inplace=True)
simpsons.set_index('Season', inplace=True)
```

```python
# Which season has the lowest number of episodes? 

simpsons['No. ofepisodes'].unique()
```

```python
simpsons = simpsons.loc[simpsons['No. ofepisodes'] != 'TBA']
```

```python
min_season = simpsons['No. ofepisodes'].min()

min_season
```

```python
simpsons.loc[simpsons['No. ofepisodes'] == min_season]
```

```python
simpsons.shape
```
